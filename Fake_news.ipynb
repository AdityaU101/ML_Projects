```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn. linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
fakenews_df = pd.read_csv("../Datasets/FakeNewsNet.csv")
```

```python
fakenews_df=fakenews_df.fillna("")
```

```python
fakenews_df['content'] = fakenews_df['title']+" "+fakenews_df["news_url"]
```

```python
fakenews_df['real'].value_counts()
```

```python
feature = fakenews_df.drop(columns='real',axis=1)
label = fakenews_df['real']
```

```python
port = PorterStemmer()
stop_words = set(stopwords.words('english'))
def stemming(content):
    stemmed_content = re.sub('^a-zA-Z',' ',content)
    stemmed_content = stemmed_content.lower().split()
    stemmed_content = [port.stem(word) for word in stemmed_content if word not in stop_words]
    stemmed_content = " ".join(stemmed_content)
    return stemmed_content
```

```python
fakenews_df['content'] = fakenews_df['content'].apply(stemming)
```

```python
features = fakenews_df['content'].values
label = fakenews_df['real'].values
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(features)

features = vectorizer.transform(features)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, label, test_size=0.2, stratify=label, random_state=2)
model=LogisticRegression()
```

```python
xtrain_prediction = model.predict(xtrain)
training_accuracy = accuracy_score(xtrain_prediction,ytrain)
print(training_accuracy)
```

```python
xtest_prediction = model.predict(xtest)
testing_accuracy = accuracy_score(xtest_prediction,ytest)
print(testing_accuracy)
```
